## Household Income Distribution Analysis

Filter the data into different income ranges and make a histogram to depict the number of households in each income range within this dataset. The first row of the income_count dataset also is irrelevent to the data itself because it just gives us descriptions of each column, not the numbers that we actally need to use for the analysis.

```python
income_count = income['S1902_C01_001E'].reset_index()                   # Putting it into a proper table form
income_count = income_count.drop(income_count.index[0])                 # Drops the first row which is irrelevent to dataset
income_count                                                            # Displays it in a proper table form
```

Even though this is now readable, it still doesn't really give us valuable information. Let's try to make household income ranges and see the number of housholds within each of these income ranges. This will be more useful data to get a better visulaization on how the income distribution looks like within this survey. The best way to approach this is to plot a histogram, which is in matplotlib library

```python
import matplotlib.pyplot as plt                                           # Importing matplotlib library to help us graph better with more tools
```

We need to create an array of household income values, so that it can be plotted using a histogram

```python
household_income = income_count['Household Income'].to_numpy()            # Making an array of values under the household income column
household_income = [int(element) for element in household_income]         # These numbers came out as strings so we are converting to integers to make histogram possible
```

Now just labelling the histogram

```python
plt.hist(household_income, bins = 100)                                    # Plotting a histogram
plt.xlabel("Household Income")                                            # Putting x-axis label
plt.ylabel("Frequency")                                                   # Putting y-axis label
plt.title("Household Income Distribution of Survey")                      # Putting title label
```

There seems to be one family that brings in 3.5M a year, and because of that and other outliers, it is harder to see the distribution through the rest of the graph because the outlier is making the rest of the graph really hard to read. So we should make 2 more graphs, one of them being only including incomes less than 400K annually and another for less than 150K annually. This will give us a better idea of how the income distributions look like closer to where almost all the surveyed families are.

```python
less_than_400K_income = [x for x in household_income if x <= 400000]
plt.hist(less_than_400K_income, bins = 100)                              # Plotting a histogram
plt.xlabel("Household Income")                                           # Putting x-axis label
plt.ylabel("Frequency")                                                  # Putting y-axis label
plt.title("Household Income Distribution of Survey Less than $400K")     # Putting title label
```

```python
less_than_150K_income = [x for x in household_income if x <= 150000]
plt.hist(less_than_150K_income, bins = 100)                              # Plotting a histogram
plt.xlabel("Household Income")                                           # Putting x-axis label
plt.ylabel("Frequency")                                                  # Putting y-axis label
plt.title("Household Income Distribution of Survey Less than $150K")     # Putting title label
```

Looking at both of these graphs, we now have a better idea of how the income distribution really looks like. We can see that most of the population surveyed are closer to the 20,000 to 50,000 annual salary range, which is less than the average household income of 74,850 nationwide. So, the demographics that are surveyed are on the lower population. That should be kept in mind when doing this analysis, especially for the Income by State analysis which is shown below. 

## Household Income by State

We can also break it down by state to see household income distribution by state. We can utilize the GEO_ID's that are given. This is US Census Data, so we can get information on the state based on it. For example, the first GEO_ID is 1600000US0103076. The state information is given in the 2 digits that follow "US" which is 01. This number corresponds to a state, which is listed in alphabetical order. So, 01 is Alabama because Alabama is the first state alphabetically.

```python
geoID = income['GEO_ID'].to_numpy()            # Making an array of values under the GEO_ID column
geoID = geoID[1:]                              # Removing 1st element because the column header got included as first value of array
```

```python
state = [int(id[9:11]) for id in geoID]        # Now to extract the 10th and 11th digit of each GEO_ID to obtain each state code. These are actually the FIPS codes for each state
income_count['State (FIP Code)'] = state       # Adding this as a new column called "State" in the income_count table
income_count                                   # Display the updated table
```

We now have combined data that we can now correspond each household to a particular state. Washington DC and Puerto Rico are included and some of the numbers are skipped. As of now, these state values are just numerical numbers but later on it will be converted to their state abbreviations.

```python
import numpy as np                             # Importing this library 
```

```python
# Make an array of all the state abbreviations in alphabetical order and which their position in the array correspond to their FIPS number
states_alphabetical = ['AL', 'AK', 'na', 'AZ', 'AR', 'CA', 'na', 'CO', 'CT', 'DE', 'DC', 'FL', 'GA', 'na', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'na', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'na', 'WI', 'WY', 'na', 'na', 'na', 'na', 'na', 'na', 'na', 'na', 'na', 'na','na', 'na', 'na', 'na', 'na', 'na', 'na', 'PR']

# Also make an array which will show which states we will actually need. This is needed because there is no data in Vermont and West Virginia and some numbers like 3 and 14 are not FIP numbers that are associated with any state.
states_unique = list(set(state))                                                                                                      # Make a new array that removes duplicates of "states" array   
states_unique = np.array(states_unique)                                                                                               # Convert it into a type of array which we can do math on easier, which we will do in the first line of for loop

plt.figure(figsize=(20, 6))                                                                                                           # Creating an area to plot the graph in. I am making it big so that it can fit all 50 states in comfortably

for x in (states_unique-1):                                                                                                           # Making a for loop to make a bar for each state. The reason it is states_unique-1 is because iteration starts from 0 while FIPS number starts from 1
    if states_alphabetical[x] == 'na':                                                                                                # We want to skip iterations where there is no states associated with the FIPS number
        continue
    state_abbr = states_alphabetical[x]                                                                                               # Obtain the state abbreviation so that that can be the x-axis label for that particular bar
    states_alphabetical[x] = income_count.loc[income_count['State (FIP Code)'] == (x+1), 'Household Income'].values.astype(int)       # Get the household income data from income_count dataset for that particular state 
    median = np.median(states_alphabetical[x])                                                                                        # Get the median household income for that state
    percentiles_25 = np.percentile(states_alphabetical[x], 25)                                                                        # Calculates the household income 25th percentile
    percentiles_75 = np.percentile(states_alphabetical[x], 75)                                                                        # Calculates the household income 75th percentile
    plt.bar(state_abbr, [median, median, median], yerr=[[median - percentiles_25], [percentiles_75 - median]])                        # Plot it in a bar chart with the 25th to 75th percentile bars

# Putting in labels
plt.xlabel('State')                                                                                                                   # Adding in the x-axis label
plt.ylabel('Household Income Annually ($)')                                                                                           # Adding in the y-axis label
plt.title('Median Household Income Surveyed by State (25th & 75th Percentile Shown)')                                                 # Adding in the title
plt.grid(axis = 'y')                                                                                                                  # Adding in horizontal gridlines for easier readability 
```

We now have a graph that shows the average household income surveyed by state. These only inlude the 646 households that were actually surveyed for this dataset, so this is not a completely accurate representation of real life. The colored bar shows the average household income and the thin, black "error" bar extends from the 25th percentile to the 75th percentile. The states that do not have an error bar associated with it are Alaska, Delaware, Washington DC, Hawaii and Maine. That is because in those states, there was only 1 household that was surveyed. That means that the value displayed on the graph is the household income for just that one house, that is why there are such bug disparities in them. Typically, but not always the smaller the error bar, the more confident we are with the median household income and that generally more households were surveyed.
